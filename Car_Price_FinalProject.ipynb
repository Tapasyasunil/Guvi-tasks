Data cleaning

```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
```

```python
df = pd.read_csv('cars_price.csv')
df.head()
```

```python
#converting ? to nan
df.replace("?", np.nan, inplace = True)
df.head(5)
```

```python
#converting to boolean to check missing data
missing_data = df.isnull()
missing_data.head(5)
```

```python
#counting missing values
for column in missing_data.columns.values.tolist():
    print(column)
    print(missing_data[column].value_counts())
    print("")
```

```python
#Finding average of column values and replacing the missing data for each column
```

```python
avg_norm_loss = df["normalized-losses"].astype("float").mean(axis=0)
print("Average of normalized-losses:", avg_norm_loss)
```

```python
df["normalized-losses"].replace(np.nan,avg_norm_loss,inplace=True)
```

```python
avg_bore = df["bore"].astype("float").mean(axis=0)
print("Average of Bore Values:",avg_bore)
```

```python
df["bore"].replace(np.nan,avg_bore,inplace=True)
```

```python
avg_stroke = df["stroke"].astype("float").mean(axis=0)
print("Average of Stroke Values:",avg_bore)
```

```python
df["stroke"].replace(np.nan,avg_stroke,inplace = True)
```

```python
avg_horsepower = df['horsepower'].astype("float").mean(axis=0)
print("Average horsepower:", avg_horsepower)
```

```python
df['horsepower'].replace(np.nan, avg_horsepower, inplace=True)
```

```python
avg_peakrpm=df['peak-rpm'].astype('float').mean(axis=0)
print("Average peak rpm:", avg_peakrpm)
```

```python
df['peak-rpm'].replace(np.nan, avg_peakrpm, inplace=True)
```

```python
df['num-of-doors'].value_counts()
```

```python
df['num-of-doors'].value_counts().idxmax()
```

```python
#replace the missing 'num-of-doors' values by the most frequent
df['num-of-doors'].replace(np.nan,"four",inplace=True)
```

```python
#dropping all rows without pricew data
# simply drop whole row with NaN in "price" column
df.dropna(subset=["price"],axis=0,inplace=True)

# reset index, because we droped two rows
df.reset_index(drop=True,inplace=True)
```

```python
df.head()
```

```python
df.dtypes
```

```python
#converting dtypes into proper format
df[["bore", "stroke"]] = df[["bore", "stroke"]].astype("float")
df[["normalized-losses"]] = df[["normalized-losses"]].astype("int")
df[["price"]] = df[["price"]].astype("float")
df[["peak-rpm"]] = df[["peak-rpm"]].astype("float")
```

```python
df.dtypes
```

```python
#data standardization
df.head()
```

```python
df["city-L/100km"] = 235 / df["city-mpg"]

df.head()
```

```python
df["highway-mpg"] = 235/df["highway-mpg"]
df.rename(columns={'"highway-mpg"':'highway-L/100km'}, inplace=True)
```

```python
#Data normalization
# replace (original value) by (original value)/(maximum value)
df["length"] = df["length"]/df["length"].max()
df["width"] = df["width"]/df["width"].max()
df['height'] = df['height']/df['height'].max()
df[["length","width","height"]].head()
```

```python
#binning the numericals
df["horsepower"] = df["horsepower"].astype(int,copy=True)
```

```python
#We will use the following function to plot the data:
def PlotPolly(model, independent_variable, dependent_variabble, Name):
    x_new = np.linspace(15, 55, 100)
    y_new = model(x_new)

    plt.plot(independent_variable, dependent_variabble, '.', x_new, y_new, '-')
    plt.title('Polynomial Fit with Matplotlib for Price ~ Length')
    ax = plt.gca()
    ax.set_facecolor((0.898, 0.898, 0.898))
    fig = plt.gcf()
    plt.xlabel(Name)
    plt.ylabel('Price of Cars')

    plt.show()
    plt.close()
```

```python
x = df['highway-mpg']
y = df['price']
#Let's fit the polynomial using the function polyfit,then use the function to display the polynomial function.
f = np.polyfit(x, y, 3)
p = np.poly1d(f)
print(p)
```

```python
PlotPolly(p, x, y, 'highway-mpg')
```

```python
#Create 11 order polynomial model with the variables x and y from above
f1 = np.polyfit(x, y, 11)
p1 = np.poly1d(f1)
print(p)
PlotPolly(p1,x,y, 'Length')
```

```python
#Use the module Pipeline to create a pipeline and also use StandardScaler as a step in our pipeline
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
```

```python
#Input the list as an argument to the pipeline constructor
pipe=Pipeline(Input)
pipe
```

```python
#Normalize the data, perform a transform and produce a prediction  simultaneously
ypipe=pipe.predict(Z)
ypipe[0:4]
```
